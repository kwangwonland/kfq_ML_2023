```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
```

```python
iris=load_iris()
```

```python
data=iris.data
label=iris.target
```

```python
model = DBSCAN(eps=0.6,
               min_samples=8,
               metric='euclidean')
pred = model.fit_predict(data)

df = pd.DataFrame(data,columns=iris.feature_names)
df['label'] = label
df['dbscan'] = pred
print(df.groupby('label')['dbscan'].value_counts())
plt.scatter(data[:,2],data[:,3],c=label)
plt.scatter(data[:,2],data[:,3],c=pred,s=2)
plt.show()
```

```python
x,y = make_circles(n_samples=1000,
                  shuffle=True,
                  noise=0.05,
                  factor=0.5)
plt.scatter(x[:,0],
            x[:,1],
            c=y)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(x)
plt.scatter(x[:,0],x[:,1],c=kmeans.labels_)
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
```

```python
gm = GaussianMixture(n_components=2)
gm.fit(x)
gm_labels = gm.predict(x)
plt.scatter(x[:,0],x[:,1],c=gm_labels)
plt.show()
```

```python
dbscan = DBSCAN(eps=0.2,
                min_samples=10,
                metric='euclidean')
d_labels = dbscan.fit_predict(x)
plt.scatter(x[:,0],x[:,1],c=d_labels)
plt.show()
```
