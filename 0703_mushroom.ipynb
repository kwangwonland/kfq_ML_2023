```python
data = df.drop(0,axis=1)
for col in data.columns:
  le = LabelEncoder()
  data[col] = le.fit_transform(data[col])
le = LabelEncoder()
label = le.fit_transform(df[0])

m1 = KNeighborsClassifier()
m2 = DecisionTreeClassifier()
m3 = LogisticRegression()
for model in [m1,m2,m3]:
  print(model.__class__.__name__)
  s = cross_val_score(model,data,label,cv=5)
  print(np.mean(s))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
s=cross_val_score(model,data,label,cv=7)
print(np.mean(s))
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
data_ss=ss.fit_transform(data)
```

```python

m1 = KNeighborsClassifier()
m2 = DecisionTreeClassifier()
m3 = LogisticRegression()
m4=RandomForestClassifier()
for model in [m1,m2,m3,m4]:
  print(model.__class__.__name__)
  s = cross_val_score(model,data_mm,label,cv=7)
  print(np.mean(s))
```

```python
model=RandomForestClassifier(max_depth=8)
s=cross_val_score(model,data,label,cv=10)
print(np.mean(s))
```

```python
depths=range(5,15)
cvs=range(5,10)

for depth in depths:
    for cv in cvs:
        model=RandomForestClassifier(max_depth=depth)
        s=cross_val_score(model,data,label,cv=cv)
        print('depth:',depth, 'cv:',cv)
        print(np.mean(s))
```

```python
depths=range(5,20)
cvs=range(5,10)

for depth in depths:
    for cv in cvs:
        model=DecisionTreeClassifier(max_depth=depth)
        s=cross_val_score(model,data_mm,label,cv=cv)
        print('depth:',depth, 'cv:',cv)
        print(np.mean(s))
```
